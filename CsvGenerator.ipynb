```python
import pandas as pd
import numpy as numpy
import os
    
```

```python
#vocabulary creation
#create necessary vocabulary for later use\
directory_name="./pins-face-recognition/"
dirs = sorted(os.listdir(directory_name))
name_to_classid = {d:i for i,d in enumerate(dirs)}
classid_to_name = {v:k for k,v in name_to_classid.items()}
num_classes = len(name_to_classid)
print("number of classes: "+str(num_classes))
```

```python
# read all directories
img_paths = {c:[directory_name+directory + "/" + img for img in sorted(os.listdir(directory_name+directory))] 
             for directory,c in name_to_classid.items()}
# retrieve all images
all_images_path = []
for img_list in img_paths.values():
    all_images_path += img_list
# map to integers
path_to_id = {v:k for k,v in enumerate(all_images_path)}
id_to_path = {v:k for k,v in path_to_id.items()}
```

```python
listall = list(img_paths.items())
```

```python
# img_paths
```

```python
listofimages=[]
classes=[]
for i in listall:
    listofimages.append(i[1])
    classes.append([i[0]]*len(i[1]))
```

```python
# celeb_classes[9000:]
```

```python
flatten = lambda l: [item for sublist in l for item in sublist]
```

```python
images=[]
celeb_classes=[]
images=flatten(listofimages)
celeb_classes= flatten(classes)
assert len(images) ==len(celeb_classes)
```

```python
dataframe = pd.DataFrame()
```

```python
dataframe["images"]=images
dataframe["class"] = celeb_classes
```

```python
dataframe.to_csv("dataset.csv",index=False)
```
